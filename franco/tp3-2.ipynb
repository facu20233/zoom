```python
import numpy as np

def jacobi(A, b, x0, max_iter, eps):
    n = len(x0)
    x1 = np.zeros(n)
    iter = 0
    error = 2 * eps

    while iter < max_iter and error > eps:
        
        iter += 1
        for i in range(n):
            suma = 0 
            for j in range(n):
                if j != i:
                    suma += A[i][j] * x0[j]
            x1[i] = (b[i] - suma) / A[i][i]
        
        error = np.linalg.norm((x1 - x0), ord=np.inf)
        x0 = x1.copy()
        
    if iter >= max_iter:
        print('El método no converge en el número máximo de iteraciones')
    else:
        print('El método converge en', iter, 'iteraciones')
        print('La solución aproximada es:', x1)

# Ejemplo de uso
a = np.array([[2, -1, 0],
             [1, 6, -2],
             [4, -3, 8]])

b = np.array([4, 2, 5])

x0 = np.array([0, 0, 0], dtype=float)

jacobi(a, b, x0, 10, 0.05)
```

```python
import numpy as np

def gauss_seidel(a, b, x0, max_iter, eps):
    iter = 0
    error = 2 * eps      
    n = len(x0)
    x1 = x0.copy()  

    while (iter < max_iter and error > eps):
        iter += 1
        for i in range(n):
            suma = 0
            
            for j in range(n):
                if j != i:
                    suma += a[i][j] * x1[j]
            
            x1[i] = (b[i] - suma) / a[i][i]
            
        error = np.linalg.norm(x1 - x0, ord=np.inf)
        x0 = x1.copy() 
        
    if (iter >= max_iter):
        print('El método no converge en el número máximo de iteraciones')
    else:
        print('El método converge en', iter, 'iteraciones')
        print('La solución aproximada es:', x1)
        

# Ejemplo de uso
a = np.array([[2, -1, 0],
              [1, 6, -2],
              [4, -3, 8]])

b = np.array([4, 2, 5])

x0 = np.array([0, 0, 0])

gauss_seidel(a, b, x0, 10, 0.05)
```

```python
import numpy as np

def sor(w, A, b, x0 ,max, eps):
    iter = 0
    error = 2 * eps      
    n = len(x0)
    x1 = x0.copy()  

    while (iter < max and error > eps):
        iter += 1
        for i in range(n):
            sum = 0
            
            for j in range(i):
                sum += a[i,j]*x1[j]
            for j in range(i+1,n):
                sum += a[i,j]*x0[j]
            
            x1[i] = (1-w)*x0[i] + (b[i]-sum)*w/A[i,i]
            
        error = np.linalg.norm(x1 - x0, ord=np.inf)
        x0 = x1.copy() 
        
    if (iter >= max):
        print('no converge')
    else:
        print('converge')

# Ejemplo de uso
a = np.array([[2, -1, 0],
              [1, 6, -2],
              [4, -3, 8]])

b = np.array([4, 2, 5])

x0 = np.array([0, 0, 0])

sor(1.5, a, b, x0, 10, 0.05)
```
