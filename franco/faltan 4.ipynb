```python
import numpy as np
import matplotlib.pyplot as plt

# a) Implementación del método de interpolación de Lagrange
def lagrange(x, y, xx):
    n = len(x)
    sum = 0
    for i in range(n):
        product = y[i]
        for j in range(n):
            if j != i:
                product = product * (xx - x[j]) / (x[i] - x[j])
        sum = sum + product
    return sum  

# b) Puntos de datos
x = [1, 2.5, 3, 3.5, 4.6, 5.1]
y = [0.0, 0.4054641, 1.0986123, 1.2527630, 1.386294, 1.6094379]

# Evaluar el polinomio de interpolación en un conjunto de puntos
xx = np.linspace(1, 5.5, 100)
yy = [lagrange(x, y, xi) for xi in xx]

# c) Graficar
plt.plot(xx, yy)
plt.scatter(x, y)
plt.grid()
plt.show()
```
