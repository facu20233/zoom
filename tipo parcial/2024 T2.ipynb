```python
import numpy as np
import matplotlib.pyplot as plt

# a)
def lagrange(x, y, xx):
    n = len(x)
    sum = 0
    for i in range(n):
        product = y[i]
        for j in range(n):
            if j != i:
                product = product*(xx - x[j]) / (x[i] - x[j]) #completar
        sum = sum + product #completar
    return sum  

# b)
x = [1, 4.6, 3, 5.1, 3.5, 2.5, 3.5]
y = [1.0, 1.386294, 1.7917595, 1.6094379, 1.0986123, 0.4054641, 0.9162907, 1.2527630]

xx = np.linspace(0, 8, 100)

# yy = [lagrange(x, y, xi) for xi in xx]

# c) Graficar
plt.plot(xx, label='Interpolación de Lagrange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolación de Lagrange')
plt.legend()
plt.grid(True)
plt.show()
            
            
    
```
